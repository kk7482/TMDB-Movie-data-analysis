```python
##calculating measures of central tendencies

def calculate_central_tendencies(df, columns):
    
    central_tendencies = pd.DataFrame()

    for column in columns:
        mean = df[column].mean()
        median = df[column].median()
        mode = df[column].mode().iloc[0] 
        central_tendencies[column] = [round(mean), round(median), round(mode)]

        central_tendencies.index = ['Mean', 'Median', 'Mode']

    return central_tendencies

columns = ['budget', 'revenue', 'runtime']
measures = calculate_central_tendencies(sort_movies, columns)

measures
```

```python
def outliers(df,col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1
    
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    
    lists = df.index[(df[col] < lf) | (df[col] > uf)]
    return lists
```
